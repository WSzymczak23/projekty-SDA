```
def plot_timeseries(axis, x, y, color, xlabel, ylabel):
  axis.plot(x, y, color=color)
  axis.set_xlabel(xlabel)
  axis.set_ylabel(ylabel, color=color)
  axis.tick_params('y', colors=color)
```

```
fig, ax = plt.subplots()
plot_timeseries(ax, merged.index, merged['Interpolated'], 'r', 'time', 'co2')
ax2 = ax.twinx()
plot_timeseries(ax2, merged.index, merged['AverageTemperature'], 'b', 'time', 'Temperature')
```

```
medals = pd.read_csv('olympic.csv')
```

```
medals.head()
```

```
def groupby_column(dataframe, column):
  groupby = dataframe.groupby(column)
  return {x:groupby.get_group(x) for x in groupby.groups}
```

```
all_sports = groupby_column(medals, 'Sport')
```

```
fig, ax = plt.subplots()
ax.bar('Gymnastics',all_sports['Gymnastics']['Height'].mean())
ax.bar('Basketball',all_sports['Basketball']['Height'].mean())

ax.set_ylabel('Height')
plt.show()
```

```
fig, ax = plt.subplots()
ax.hist(all_sports['Gymnastics']['Height'], bins=20)
ax.hist(all_sports['Basketball']['Height'], bins=20)

plt.show()
```

```
fig, ax = plt.subplots()
ax.bar('Gymnastics',all_sports['Gymnastics']['Height'].mean(),
       yerr = all_sports['Gymnastics']['Height'].std())
ax.bar('Basketball', all_sports['Basketball']['Height'].mean(),
       yerr = 'Basketball', all_sports['Basketball']['Height'].std()

plt.show()
```
